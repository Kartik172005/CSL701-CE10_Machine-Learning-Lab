```python
#Name: Kartik Gawade CE10

#Task 1: Import NumPy, Pandas, Matplotlib.pyplot and Seaborn.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside notebook
%matplotlib inline
```

```python
#Task 2: Load train.csv into a Pandas DataFrame using pd.read_csv().
from google.colab import files

uploaded = files.upload()
df = pd.read_csv("Titanic-Dataset.csv")

print("Dataset Loaded Successfully!")
```

```python
#Task 3: Display first 10 rows using df.head(10).
df.head(10)
```

```python
#Task 4: Display last 5 rows using df.tail(5).
df.tail(5)
```

```python
#Task 5: Check dataset dimensions using df.shape (Expected: 891 x 12).
print("Shape of Dataset:", df.shape)
```

```python
#Task 6: Display dataset information using df.info().
df.info()
```

```python
#Task 7: Generate summary statistics using df.describe(); record mean, minimum and maximum values of Age and Fare.
df.describe()
print("Age Statistics")
print("Mean :", df["Age"].mean())
print("Minimum :", df["Age"].min())
print("Maximum :", df["Age"].max())

print("\nFare Statistics")
print("Mean :", df["Fare"].mean())
print("Minimum :", df["Fare"].min())
print("Maximum :", df["Fare"].max())
```

```python
#Task 8: Identify missing values using df.isnull().sum().
df.isnull().sum()
```

```python
#Task 9: Fill missing Age values with the median.

median_age = df["Age"].median()

df["Age"] = df["Age"].fillna(median_age)

print("Missing Age values filled successfully.")
```

```python
#Task 10: Filter surviving passengers (Survived == 1).
survivors = df[df["Survived"] == 1]

survivors.head()
```

```python
#Task 11: Convert Fare column to a NumPy array.
fare_array = df["Fare"].to_numpy()

print(fare_array)
```

```python
#Task 12: Calculate maximum and average Fare using np.max() and np.mean().
print("Maximum Fare :", np.max(fare_array))
print("Average Fare :", np.mean(fare_array))
```

```python
#Task 13: Plot histogram of Age.
plt.figure(figsize=(8,5))

plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
#Task 14: Plot bar chart of Survival Count.
plt.figure(figsize=(6,4))

sns.countplot(x="Survived", data=df)

plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")

plt.show()
```

```python
#Task 15: Plot scatter graph of Age vs Fare.
plt.figure(figsize=(5,4))

plt.scatter(df["Age"], df["Fare"], color="red", alpha=0.6)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```
